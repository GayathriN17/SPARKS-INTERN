```python
data['State'].value_counts().head(10)

data_state= data.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
data_state.head(10)
```

```python
data_state1=data_state.sort_values('Sales')

data_state1[['Sales']].plot(kind = 'line', figsize = (15,4))
plt.title('State wise sales Analysis', fontsize = 20)
plt.ylabel('sales')
plt.xlabel('States')
plt.show()
```

```python
data_state1=data_state.sort_values('Discount')

data_state1[['Discount']].plot(kind = 'line', figsize = (15,4))
plt.title('State wise Discount Analysis', fontsize = 20)
plt.ylabel('Discount')
plt.xlabel('States')
plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.pie(data_sub_category['Profit'], labels = data_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Profit Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```
